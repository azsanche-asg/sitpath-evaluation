## Plot satisfaction vs ΔADE

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
dfs = []
for fname in ['controllability_avoid.csv', 'controllability_keep.csv', 'controllability_tempo.csv']:
    path = Path('tables') / fname
    if path.exists():
        dfs.append(pd.read_csv(path))
summary = pd.concat(dfs, ignore_index=True)
display(summary)
plt.figure(figsize=(5,3))
plt.bar(summary['strategy'], summary['constraint_rate'], label='constraint rate')
plt.plot(summary['strategy'], summary['delta_ADE'], 'ro-', label='ΔADE (m)')
plt.ylabel('Rate / meters')
plt.legend()
plt.tight_layout()
plt.show()
```
